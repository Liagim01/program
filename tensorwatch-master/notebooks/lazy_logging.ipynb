## Basic Concept
So what is lazy logging? Here we don't actually log any thing explicitly. Instead we have opportunity to observe variables. Observing variables is very cheap so we can observe all the variables if we want to. At the runtime, you can specify arbitrary *lambda expression* that we execute on the variables being observed. The result of this lambda expression is a stream that you can then render in to visualizer of choice. That's pretty much it!

While `sum_lazy.py` script is running, you can execute [this notebook](https://github.com/microsoft/tensorwatch/blob/master/notebooks/lazy_logging.ipynb).

```python
%matplotlib notebook
import tensorwatch as tw
```

Now we create the stream by specifying a lambda expression that would operate on variables we are observing. Below we simply ask to sum all values of array and sent the result as stream.

```python
client = tw.WatcherClient()
stream = client.create_stream(expr='lambda d: np.sum(d.weights)')
```

```python
line_plot = tw.Visualizer(stream, vis_type='line')
line_plot.show()
```
